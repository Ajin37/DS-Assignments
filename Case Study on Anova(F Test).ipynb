### Does each Zone Generate Same Amount of Sales

```python
# H0: Each zone generate same amount of sales
# Ha: Each zone does not generate same amount of sales
```

```python
alpha=0.05
fvalue, pvalue = stats.f_oneway(df['Zone - A'], df['Zone - B'], df['Zone - C'],df['Zone - D'])
print(fvalue,pvalue)
```

Here the pvalue is less than the alpha thus we reject the null hypothesis
#### And there is sufficient evidence to say that all the zones did not generate same amount of sales
