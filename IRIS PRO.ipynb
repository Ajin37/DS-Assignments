### Visualizations

### 1. Pairplot

```python
sns.pairplot(df,hue='Classification',height=3)
```

### 2.ScatterPlot b/w Petal Length and Width

```python
x=df['PL']
y=df['PW']
h=df['Classification']

plt.figure(figsize=(16,9))
plt.title('Comparison between petal width and length on the basis of species')
sns.scatterplot(x,y, hue=h, s= 50);
```

-Due to the grouping of setosa'a data points, given the length and width of a petal we can confirm if it is setosa or not

### 3.ScatterPlot b/w Sepal Width and Length

```python
x=df['SL']
y=df['SW']
h=df['Classification']

plt.figure(figsize=(16,9))
plt.title('Comparison between sepal width and length on the basis of species')
sns.scatterplot(x,y, hue=h, s= 50);
```

-Setosa is the shorter Irsis, but it also has more width than others<br>
-Virginica tend to be longer but less broader

### 4.HeatMap

```python
plt.figure(figsize=(10,11))
sns.heatmap(df.corr(),annot=True)
plt.plot()

```

We can see that there is indeed a strong positive correlation between petal length and width, as well as, petal length and sepal length, and sepal length with petal width
